**Numpy Indexing and Slicing of Vectors**

```python
import numpy as np
```

```python
arr = np.arange(11,21)
arr
```

```python
arr[6]
```

```python
arr[1:5]
```

**Numpy Indexing and Slicing of Matrix**

```python
arr = np.arange(1,31).reshape(6,5)
arr
```

```python
arr[0]
```

```python
arr[1]
```

```python
arr[0,0]
```

```python
arr[5,4]
```

```python
arr[0:2,1:3]
```

```python
slice = arr[3:,3:]
slice
```

**Boolean Indexing**

```python
arr = np.arange(11,21)
arr
```

```python
bool_index = arr % 2 == 0
bool_index
```

```python
arr = arr[bool_index]
arr
```
