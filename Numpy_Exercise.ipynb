**Valid Sudoku**

```python
import numpy as np
```

```python
s = np.array([
    [5, 3, 4, 6, 7, 8, 9, 1, 2],
    [6, 7, 2, 1, 9, 5, 3, 4, 8],
    [1, 9, 8, 3, 4, 2, 5, 6, 7],
    
    [8, 5, 9, 7, 6, 1, 4, 2, 3],
    [4, 2, 6, 8, 5, 3, 7, 9, 1],
    [7, 1, 3, 9, 2, 4, 8, 5, 6],
    
    [9, 6, 1, 5, 3, 7, 2, 8, 4],
    [2, 8, 7, 4, 1, 9, 6, 3, 5],
    [3, 4, 5, 2, 8, 6, 1, 7, 9]
])
```

```python
import numpy as np
a = np.sum(s,axis=1) 
for i in a:
    if i !=45:
        print("Sudoku is not valid")
else:
    print("Fow rows it is valid")
```

```python
b = np.sum(s,axis=0)
for i in b:
    if i != 45:
        print("Sudoku is not valid")
else:
    print("Fow columns it it valid")
```

```python
#Using For Loop
for i in range(0,9,3):
    for j in range(0,9,3):
        n = s[i:i+3,j:j+3]
        print(n.sum())
        
```

**General Questions**

```python
import numpy as np

# Columns: [Age, Math Marks, Science Marks]
data = np.array([
    [18, 85, 78],   # Student 1
    [19, 92, 88],   # Student 2
    [17, 76, 95],   # Student 3
    [18, 65, 70],   # Student 4
    [20, 90, 85]    # Student 5
])
```

```python
#Get the shape of the matrix.
data.shape #Here we will not use brackets because shape is an instance attribute
```

```python
#Find the average age of students.
np.mean(data[:,0])
```

```python
#Extract math marks of all students.
data[:,1]
```

```python
#Find the highest science marks.
np.max(data[:,2])
```

```python
#Get details of the students who scored more than 90 in maths.
data[data[:,1] > 90]
```

```python
#Increase math marks of all students by 5.
data[:,1] += 5 # Arrays has broadcasting so we directly add 5
data
```

```python
#Get data of students who scored at least 80 in both subjects.
data[(data[:,1] >= 80) & (data[:,2] >= 80)]
```
